## Descrição: Crie uma classe base Animal com um método que exibe o som que o animal faz. Crie subclasses Cachorro e Gato que sobrescrevam o método da classe base para exibir o som específico que fazem.

```python

class Animal:
  def __init__(self, tipo):
    self.tipo = tipo
  def som(self):
    print(f'{self.tipo} faz um som')

class Cachorro(Animal):
    def som(self):
        print(f"{self.tipo} faz: latido")

class Gato(Animal):
    def som(self):
        print(f"{self.tipo} faz: miado")


meu_cachorro = Cachorro("Rex")
meu_gato = Gato("Mimi")

meu_cachorro.som()
meu_gato.som()
```

```python
class Pessoa:
  def __init__(self,nome, idade, cidade):
    self.nome = nome
    self.idade = idade
    self.cidade = cidade
  def salvar(self, arquivo):
      with open(arquivo, 'w') as file:
          file.write(f"{self.nome}\n")
          file.write(f"{self.idade}\n")
          file.write(f"{self.cidade}\n")
  @classmethod
  def carregar(cls, arquivo):
      with open(arquivo, 'r') as file:
          nome = file.readline().strip()
          idade = file.readline().strip()
          cidade = file.readline().strip()
      return cls(nome, idade, cidade)


# Criando um objeto da classe Pessoa
p1 = Pessoa("João", 30, "Rio de Janeiro")
# Salvando os dados da pessoa em um arquivo
p1.salvar('pessoa.txt')

# Carregando os dados do arquivo e criando um novo objeto
p2 = Pessoa.carregar('pessoa.txt')
print(p2.nome, p2.idade, p2.cidade)

```

```python
numero = int(input('Digite um número encontrar a sequência de Fibonacci: '))
fibonacci = [0,1,1]
soma = 0
contador = 2
primeiro, segundo = 1,1
if numero == 1:
  print(f'O número da sequência é {fibonacci}')
elif numero == 0:
  print(f'O número da sequência é {fibonacci[0]}')
elif numero < 0:
  print(f'Você precisa digitar um número positivo!')
else:
  while True:
    soma = primeiro + segundo
    primeiro, segundo = segundo, soma
    fibonacci.append(soma)
    contador += 1
    if fibonacci[contador] >= numero:
      if fibonacci[contador] == numero:
        break
      elif fibonacci[contador] > numero:
        fibonacci.pop()
        break
  print(fibonacci)

```

```python
class Veiculo:
  def __init__(self,marca, modelo, ano):
    self.marca = marca
    self.modelo = modelo
    self.ano = ano

  def exibir_detalhes(self):
    print(f'A marca do carro é {self.marca}, o modelo do carro é {self.modelo} e o ano dele é {self.ano}')

class Carro(Veiculo):
  def __init__(self,marca, modelo, ano, numero_de_portas):
    super().__init__(marca, modelo, ano)
    self.numero_de_portas = numero_de_portas

  def exibir_detalhes(self):
    super().exibir_detalhes()
    print(f'O carro tem {self.numero_de_portas} portas')

class Moto(Veiculo):
    def __init__(self,marca, modelo, ano, tipo_de_guidao):
      super().__init__(marca, modelo, ano)
      self.tipo_de_guidao = tipo_de_guidao

    def exibir_detalhes(self):
      super().exibir_detalhes()
      print(f'A moto tem o guidão {self.tipo_de_guidao}')

carro = Carro('Gol', 'GV20', 2000, 4)
moto = Moto('Honda', 'CB500', 2010, 'Trianguloar')
carro.exibir_detalhes()
moto.exibir_detalhes()
```

```python
def frequencia_palavras (frase, organizar = False):
  frase = frase.replace(",", "").lower()
  frase = frase.split()
  if organizar == True:
    frase = sorted(frase)
  dicionario = {}
  for palavra in frase:
    if palavra not in dicionario:
      dicionario[palavra] = 1
    else:
      dicionario[palavra] += 1
  return dicionario

frase = input('Digite uma frase para iniciar: ')

print(f'Cada palavra aparece com essa frequência {frequencia_palavras(frase)}, as palavras em ordem crescente ficam {frequencia_palavras(frase, True)}')
```

```python
class FormaGeometrica:
  def calcular_area(self):
      raise NotImplementedError("Este método deve ser implementado pelas subclasses")

class Retangulo(FormaGeometrica):
  def __init__(self, largura, altura):
    self.largura = largura
    self.altura = altura
    self.area = self.largura * self.altura
  def calcular_area(self):
    print(f'A área do retângulo é {self.area}')
class Circulo(FormaGeometrica):
  def __init__(self, raio):
    self.raio = raio
    pi = 3.14
    self.area = self.raio **2 * pi
  def calcular_area(self):
    print(f'A área do círculo é {self.area}')
class Triangulo(FormaGeometrica):
  def __init__(self, a, b, c):
    self.a = a
    self.b = b
    self.c = c
  def calcular_area(self):
    calcular_s = (self.a + self.b + self.c) / 2
    area = (calcular_s*(calcular_s - self.a)* (calcular_s - self.b)* (calcular_s - self.c))**(1/2)
    print(f'A área do triângulo é {area}')

escolha = input('Escolha qual das três figuras quer realizar o cálculo da área (Retangulo, Circulo ou Triangulo): ').lower()

if escolha == "retangulo":
  largura = int(input('Digite a largura: '))
  comprimento = int(input("Digite o comprimento: "))
  retangulo = Retangulo(largura,comprimento)
  retangulo.calcular_area()
elif escolha == "circulo":
  raio = int(input('Digite o raio: '))
  circulo = Circulo(raio)
  circulo.calcular_area()
elif escolha == "triangulo":
  a = int(input('Digite o lado a: '))
  b = int(input("Digite o lado b: "))
  c = int(input("Digite o lado c: "))
  triangulo = Triangulo(a,b,c)
  triangulo.calcular_area()
else:
  print('Não tenho essa alternativa, por favor, tente novamente')
```

```python
class ContaBancaria:
  def __init__(self, titular, saldo):
    self.titular = titular
    self.saldo = saldo
  def depositar(self, saldo):
    self.saldo = self.saldo + saldo
    print(f'Seu saldo é {self.saldo}, Sr(a){self.titular}')
  def sacar(self, sacar):
    if self.saldo >= sacar:
      self.saldo = self.saldo - sacar
      print(f'Agora seu saldo é {self.saldo}, Sr(a){self.titular}')
    else:
      print("Seu saldo é inferior ao valor que deseja sacar, seu saldo atual é de {}".format(self.saldo))
  def transferir(self, valor, conta_destino):
      if valor <= self.saldo:
          self.saldo -= valor
          conta_destino.saldo += valor
          print(f"{self.titular} transferiu {valor} para {conta_destino.titular}. Saldo atual: {self.saldo}")
      else:
          print(f"{self.titular} tentou transferir {valor}, mas o saldo é insuficiente. Saldo atual: {self.saldo}")

conta1 = ContaBancaria("Alice", 500)
conta2 = ContaBancaria("Bob", 300)
conta1.depositar(200)
conta2.depositar(100)
conta1.sacar(100)
conta2.sacar(500)
conta1.transferir(300, conta2)
conta2.transferir(50, conta1)


```

```python
from datetime import datetime



class CarteiraDigital:
  def __init__(self, dono, saldo):
    self.dono = dono
    self.saldo = saldo
    self.historico = []
  def depositar(self, valor):
    self.saldo += valor
    resposta = f"Você depositou R$ {valor}, Sr(a) {self.dono}, seu saldo foi para R$ {self.saldo}, isso ocorreu na data e hora:{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"
    self.historico.append(resposta)
  def sacar(self, valor):
    if self.saldo > 0:
      self.saldo -= valor
      resposta = f"Você sacou R$ {valor}, Sr(a) {self.dono}, seu saldo foi para R$ {self.saldo}, isso ocorreu na data e hora:{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"
      self.historico.append(resposta)
    else:
      print(f'Você não tem saldo suficiente, seu saldo é: {self.saldo}')
  def exibir_historico(self):
    for i in self.historico:
      print(i)


pessoa1 = CarteiraDigital('Ana', 1200)
pessoa2 = CarteiraDigital('João', 0)
pessoa3 = CarteiraDigital('Alberto', 5000)

pessoa1.depositar(1000)
pessoa1.sacar(500)
pessoa1.exibir_historico()


pessoa2.depositar(5000)
pessoa2.sacar(10)
pessoa2.exibir_historico()


pessoa3.depositar(40)
pessoa3.sacar(400)
pessoa3.exibir_historico()
```
